- A simple GET request

```python
from requests import get
r = get('https://httpbin.org/get')
r.status_code
```

```python
import urllib3
http = urllib3.PoolManager()

URL = 'https://docs.snowflake.com/en/release-notes/requirements#recommended-client-versions'
resp =http.request('GET', URL)
resp.status
```

```python
resp.data
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(resp.data.decode(), "html.parser")
soup.title
```

```python
soup
```

```python
import pandas as pd
tables = pd.read_html(str(soup), encoding='utf-8')
tables
```

```python
ver_table = tables[0]
ver_table
```

```python
ver_table.to_records(index=False).tolist()
```

```python
import requests

URL = 'https://docs.snowflake.com/en/release-notes/requirements#recommended-client-versions'
resp = requests.get(URL)
resp.status_code
```

```python
resp.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(resp.text, "html.parser")
soup
```

```python
import pandas as pd
from io import StringIO
tables = pd.read_html(StringIO(str(soup)), encoding='utf-8')
tables
```

```python
ver_table = tables[0]
ver_table
```

```python
ver_table.to_records(index=False).tolist()
```
